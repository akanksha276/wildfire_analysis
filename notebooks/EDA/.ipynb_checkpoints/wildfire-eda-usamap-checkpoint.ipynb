```python
# Connect to the sqlite db file and retrieve data as Pandas data frame.
cnx = sqlite3.connect('/kaggle/input/188-million-us-wildfires/FPA_FOD_20170508.sqlite')
sql = "select * from fires"
df = pd.read_sql_query(sql, cnx)
```

```python
df.head()
```

```python
#I ADDED
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['FIRE_SIZE_SCALED'] = scaler.fit_transform(df[['FIRE_SIZE']])
```

```python
# Check NA values in features of interest.
df_interest = df[["FIRE_YEAR","DISCOVERY_DOY","DISCOVERY_TIME","CONT_DOY","CONT_TIME","STAT_CAUSE_CODE","STAT_CAUSE_DESCR","FIRE_SIZE","FIRE_SIZE_SCALED","LATITUDE","LONGITUDE","STATE","COUNTY","FIPS_NAME","DISCOVERY_DATE"]]
df_interest.isna().sum()
```

County feature has many NA's. This needs to be handled for the second question. 

```python
# Check and remove duplicate rows.
print(f"Number of removed duplicate rows: {df_interest.duplicated().sum()}")
df_interest = df_interest.drop_duplicates()
```

```python
# Check data values
print(df_interest.STATE.unique())
print(df_interest.COUNTY.unique())
```

County values have numeric FIPS codes and also word format. So, instead of selecting COUNTY feature, FIPS_NAME is selected for the second question.(Also some counties in COUNTY feature are misstyped like Maui and Mauii)

```python
# Convert Julian date format into Gregorian date format.
df_interest["combined_date_dis"] = df_interest["FIRE_YEAR"]*1000 + df_interest["DISCOVERY_DOY"]
df_interest["combined_date_dis"] = pd.to_datetime(df_interest["combined_date_dis"], format = "%Y%j")
df_interest = df_interest.set_index("combined_date_dis")
df_interest.head()
```

```python
#I ADDED
import folium
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.cluster import MiniBatchKMeans
from geopy.distance import great_circle
```

```python
#I ADDED
# define the minimum number of fires in a cluster and the maximum distance between fires in a cluster
min_cluster_size = 10
max_cluster_distance = 1000 # in meters
```

```python
#I ADDED
# create a list of fire coordinates
coords = df_interest[['LATITUDE', 'LONGITUDE']].values
```

```python
#I ADDED
# compute the clusters using DBSCAN
# db = DBSCAN(eps=max_cluster_distance/1000, min_samples=min_cluster_size/2, algorithm='ball_tree', metric='haversine').fit(np.radians(coords))
# db = DBSCAN(eps=max_cluster_distance/1000, min_samples=min_cluster_size/2, algorithm='ball_tree', metric='haversine', memory='./').fit(np.radians(coords))
kmeans = MiniBatchKMeans(n_clusters=1000, batch_size=10000).fit(np.radians(coords))

# df_interest['cluster'] = db.labels_
df_interest['cluster'] = kmeans.labels_
```

```python
#I ADDED
# create a map centered on the USA
map = folium.Map(location=[37, -102], zoom_start=4)

# loop through the clusters and add markers for each cluster with the size indicated by the marker color
for cluster in df_interest['cluster'].unique():
    if cluster != -1:
        cluster_df = df_interest[df_interest['cluster']==cluster]
        center_lat, center_lon = cluster_df[['LATITUDE', 'LONGITUDE']].mean()
        size = cluster_df['FIRE_SIZE_SCALED'].sum()
        folium.CircleMarker(location=[center_lat, center_lon], 
                            radius=size, 
                            color='red', 
                            fill=True, 
                            fill_color='red', 
                            fill_opacity=0.5).add_to(map)

# display the map
map
```

```python
# Examine monthly and yearly counts of fires.
monthly_trend = df_interest.groupby(pd.Grouper(freq='M')).count()[["FIRE_YEAR"]]
monthly_trend.columns = ["Number of fires (Monthly)"]
monthly_trend.plot()

yearly_trend = df_interest.groupby(pd.Grouper(freq='Y')).count()[["FIRE_YEAR"]]
yearly_trend.columns = ["Number of fires (Yearly)"]
yearly_trend.plot()
```
